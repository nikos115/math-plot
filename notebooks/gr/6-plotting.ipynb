## Ας ανακεφαλαιώσουμε!

Εισάγουμε τις απαραίτητες βιβλιοθήκες για μαθηματικούς υπολογισμούς και απεικόνιση

```python
import matplotlib.pyplot as plt
from math import pow, log, log10, sqrt
```

Ορίζουμε μαθηματικές συναρτήσεις

```python
def square(x):
    return pow(x, 2)


def cube(x):
    return pow(x, 3)


def ln(x):
    if x <= 0:
        return None
    else:
        return log(x)


def logarithm(x):
    if x <= 0:
        return None
    else:
        return log10(x)


def square_root(x):
    if x < 0:
        return None
    else:
        return sqrt(x)
```

Δημιουργούμε μια λίστα με τιμές

```python
x_values = range(-20, 21, 1)
```

Υπολογίζουμε τα y για τα αντίστοιχα x

```python
y_square = list(map(square, x_values))
y_cube = list(map(cube, x_values))
y_ln = list(map(ln, x_values))
y_logarithm = list(map(logarithm, x_values))
y_square_root = list(map(square_root, x_values))
```

Απεικονίζουμε το τετράγωνο και τον κύβο

```python
plt.plot(x_values, y_square, label='square')
plt.plot(x_values, y_cube, label='cube')

plt.legend()

plt.show()
```

Απεικονίζουμε το λογάριθμο και την τετραγωνική ρίζα

```python
plt.plot(x_values, y_ln, label='ln')
plt.plot(x_values, y_logarithm, label='logarithm')
plt.plot(x_values, y_square_root, label='square_root')

plt.legend()

plt.show()
```
