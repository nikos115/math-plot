# Math with Python - Plotting math functions

Import required modules for calculating basic math and plotting

```python
import matplotlib.pyplot as plt
import numpy as np
from math import pow, log, log10, sqrt
```

Define math functions

```python
def square(x):
    return pow(x, 2)


def cube(x):
    return pow(x, 3)


def ln(x):
    if x <= 0:
        return None
    else:
        return log(x)


def logarithm(x):
    if x <= 0:
        return None
    else:
        return log10(x)


def square_root(x):
    if x < 0:
        return None
    else:
        return sqrt(x)
```

Create a list of x values

```python
x_values = np.arange(-20, 20.1, 0.5)
```

Calculate y values for the respective x values

```python
y_square = list(map(square, x_values))
y_cube = list(map(cube, x_values))
y_ln = list(map(ln, x_values))
y_logarithm = list(map(logarithm, x_values))
y_square_root = list(map(square_root, x_values))
```

Plot square and cube functions

```python
plt.plot(x_values, y_square, label='square')
plt.plot(x_values, y_cube, label='cube')

plt.legend()

plt.show()
```

Plot logarithmic and square root functions

```python
plt.plot(x_values, y_ln, label='ln')
plt.plot(x_values, y_logarithm, label='logarithm')
plt.plot(x_values, y_square_root, label='square_root')

plt.legend()

plt.show()
```
